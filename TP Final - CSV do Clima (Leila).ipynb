```python
#Descrição
#Pro trabalho fiz modular com a abertura dos arquivos e as funções que calcula as condições pedidas no enunciado
#Usei os comandos ifs para imprimir o calculo de cada comando (se as condições forem satisfeitas)

#entrada de dados
cond = input()
inic_year = int(input())
final_year = int(input())

#funcao principal
def main():
    inic_year = read.inic_year
    final_year = read.final_year

#arquivo temperatura
def load_temperature():
    with open("temp.csv", "r") as file_temperature:
        next(file_temperature) #remove a 1° linha do arquivo
        temperature_data = {} #criar o diretorio 
        for line in file_temperature: #ordena pela ",", e no dicionario o ano eh a chave
            year = int(line.strip().split(",")[0])
            temperature_data[year] = float(line.strip().split(",")[1])
    return temperature_data

#arquivo de csv
def load_co2():
    with open("co2.csv", "r") as file_co2:
        co2_data = {} #cria o diretorio 
        for line in file_co2:
            #considera no arquivo apenas World
            x = str(line.strip().split(",")[0])
            if x == "World": 
                year = int(line.strip().split(",")[2])
                co2_data[year] = float(line.strip().split(",")[3])
    return co2_data

#retorna os arquivos nas funcoes
temperature_data = load_temperature()
co2_data = load_co2()

#funcoes das condicoes
#criei as funcoes anomalia e emissoes para temperature e co2
#elas irao criar listas e calcular o intervalo de anos digitados pelo usuario
def anomalia(temperature_data,inic_year,final_year):
    LT = [] #cria a lista da condicao 
    for ano, anomalia in temperature_data.items(): #for do ano e anomalia no dicionario
        if inic_year <= ano <= final_year: #se ano entre o intervalo
            LT.append(anomalia) #apenda na lista 
    return LT #retorna 

def emissoes(co2_data,inic_year,final_year):
    LC = []
    for ano, co2 in co2_data.items():
        if inic_year <= ano <= final_year:
            LC.append(co2)
    return LC

#funcao para a media movel
def medias(co2_data,inic_year,final_year,q_media,intervalo,passo):
    medias = [] #cria a lista 
    q_media = int(input()) #pede a quantidade de medias que quer calcular
    intervalo = int(input()) #quantos intervalos quer 
    passo = int(input()) #quantos passos vai dar dentro do intervalo 
    for i in range(q_media): #calcula a media movel pela soma do co2 do indice ao passo (dado o intervalo)
        media_movel = sum(emissoes[i * passo: intervalo + (i * passo)])/ intervalo
        medias.append(media_movel) #appenda na lista
    return medias #retorna
    
#ifs das condicoes
anomalia = anomalia(temperature_data,inic_year,final_year)    
emissoes = emissoes(co2_data,inic_year,final_year)

if ("LT" in cond):
    for value in anomalia:
        print("%.8f" % value)
        
if ("LC" in cond):
    for value in emissoes:
        print("%.2f" % value)
        
if ("IT" in cond):
    print("%.8f" % min(anomalia))
    
if ("IC" in cond):
    print("%.2f" % min(emissoes))
    
if ("ST" in cond):
    print("%.8f" % max(anomalia))
    
if ("SC" in cond):
    print("%.2f" % max(emissoes))   
    
#utilizei as funcoes de anomalia e emissoes ja feitas e acrescentei as medias no if
if ("MT" in cond):
    media_temp = sum(anomalia)/len(anomalia)
    print("%.4f" % (media_temp))
    
if ("MC" in cond): 
    media_co2 = sum(emissoes)/len(emissoes)
    print("%.4f" % media_co2)
    
#se a condicao for satisfeita, chama a funcao "medias" (que calcula a media movel)   
if ("MM" in cond):
    medias = medias(co2_data,inic_year,final_year,q_media,intervalo,passo)
    for value in medias:
        print("%.4f" % value)

#A FAZER
if ("CC" in cond): 
    print("{:.4f}" % correlacao)
```
